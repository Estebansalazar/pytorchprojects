```
import torch
```

```
#transformamos este vector en una matriz de 2x3 (view(2,3))
mat_a=torch.tensor([8,7,5,4,2,2]).view(2,3)
mat_a
```

```
#transformamos este vector en una matriz de 3x2 (view(3,2))

mat_b=torch.tensor([3,4,3,-2,4,-2]).view(3,2)
mat_b
```

```
#multiplicame la matriz a con la b
torch.matmul(mat_a,mat_b)
```
